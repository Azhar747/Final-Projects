## Removing Outliers with Zscore Method

```python
from scipy.stats import zscore
z=np.abs(zscore(data))
z.shape
```

```python
threshold=3
print(np.where(z>3))
```

```python
data_new=data[(z<3).all(axis=1)]
print(data.shape)
print(data_new.shape)
```

```python
Data_loss=((18249-18118)/18249)*100
```

```python
Data_loss
```

```python
data_new.skew()
```

Separating Features and Target

```python
X=data_new.drop("AveragePrice",axis=1)
y=data_new["AveragePrice"]
```

## Feature Scaleing

```python
from sklearn.preprocessing import MinMaxScaler
mms=MinMaxScaler()
```

```python
X_mms=mms.fit_transform(X)
```

## Train Test Split

```python
from sklearn.model_selection import train_test_split
```

```python
 X_train, X_test, y_train, y_test = train_test_split(X_mms, y, test_size=0.33, random_state=42)
```

```python
X_train
```

```python
y
```

## Linear Regression 

```python
from sklearn.linear_model import LinearRegression
lin_regg=LinearRegression()
```

```python
lin_regg.fit(X_train,y_train)
```

```python
y_pred=lin_regg.predict(X_test)
```

```python
from sklearn import metrics
print('MAE:', metrics.mean_absolute_error(y_test, y_pred))
print('MSE:', metrics.mean_squared_error(y_test, y_pred))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```

RMSE is low lets try another model and check 

```python
metrics.r2_score(y_test, y_pred)
```

## Cross Validation of Linear Regression

```python
from sklearn.model_selection import cross_val_score
cv_score=cross_val_score(lin_regg,X,y,cv=5).mean()
cv_score
```

## Regularization

## Decition Tree Regressor 

```python
from sklearn.tree import DecisionTreeRegressor
dr=DecisionTreeRegressor()
dr.fit(X_train,y_train)
pred=dr.predict(X_test)
```

```python
print('MAE:', metrics.mean_absolute_error(y_test, pred))
print('MSE:', metrics.mean_squared_error(y_test, pred))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, pred)))
```

```python
metrics.r2_score(y_test, pred)
```

```python
from sklearn.model_selection import cross_val_score
cv_score=cross_val_score(dr,X,y,cv=5).mean()
cv_score
```
