## Now we will handle the ? values  and Categorical Columns

```python
df['Workclass'] = df['Workclass'].str.replace('?', 'Private', regex=True) ## So we have replaced ? with private which is repeated more number of times 
```

```python
df['Occupation'] = df['Occupation'].str.replace('?', 'Prof-specialty', regex=True)# So we have replaced ? with Prof-specialty which is repeated more number of times "]=df["Workclass"].replace("?","Prof-specialty")# So we have replaced ? with Prof-specialty which is repeated more number of times 
```
