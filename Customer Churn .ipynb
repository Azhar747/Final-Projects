## Encoding the Categorical Data

```python
dumm=pd.get_dummies(df["Churn"],drop_first=True)
```

```python
df=pd.concat([df,dumm],axis=1)
```

```python
df.head()
```

```python
df.drop("Churn",axis=1,inplace=True)
```

```python
df.rename({"Yes":"Churn"},axis=1,inplace=True)
```

```python
df.head()
```

```python
df.info()
```

```python
df["Churn"].value_counts()
```

## Encoding Other Column using Lable Encoder

```python
from sklearn.preprocessing import LabelEncoder
```

```python
for column in df.columns:
    if df[column].dtype==np.number:
        continue
    else:
        df[column]=LabelEncoder().fit_transform(df[column])
```

```python
df.info()
```

```python
df["Churn"].value_counts()
```

```python
X=df.drop("Churn",axis=1)
```

```python
y=df["Churn"]
```

## Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
scalar=StandardScaler()
```

```python
X_scalar=scalar.fit_transform(X)
```

## Performing train test split

```python
from sklearn.model_selection import train_test_split
```

```python
 X_train, X_test, y_train, y_test = train_test_split(X_scalar, y, test_size=0.33, random_state=42)
```

```python
print("Number transactions X_train dataset: ", X_train.shape)
print("Number transactions y_train dataset: ", y_train.shape)
print("Number transactions X_test dataset: ", X_test.shape)
print("Number transactions y_test dataset: ", y_test.shape)
```

```python
df["Churn"].value_counts()
```

```python
import imblearn
```

```python
from imblearn.over_sampling import SMOTE
```

```python
print("Before OverSampling, counts of label '1': {}".format(sum(y_train==1)))
print("Before OverSampling, counts of label '0': {} \n".format(sum(y_train==0)))
```

```python
sm = SMOTE(random_state=42)
X_res, y_res = sm.fit_resample(X_train, y_train)

print('After OverSampling, the shape of train_X: {}'.format(X_res.shape))
print('After OverSampling, the shape of train_y: {} \n'.format(y_res.shape))

print("After OverSampling, counts of label '1': {}".format(sum(y_res==1)))
print("After OverSampling, counts of label '0': {}".format(sum(y_res==0)))
```

## Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
Logi_reg=LogisticRegression()
```

```python
Logi_reg.fit(X_res,y_res)
```

```python
y_pred=Logi_reg.predict(X_test)
```

```python
## Logistic Regression Model Evaluation
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
```

```python
confusion_matrix(y_test, y_pred)
```

```python
print(classification_report(y_test, y_pred))
```

```python
logi_accuracy=accuracy_score(y_test, y_pred)
```

```python
logi_accuracy
```
