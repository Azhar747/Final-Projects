## Splitting the data

```python
X=df_new.drop("W",axis=1)
```

```python
y=df_new["W"]
```

```python
X.head()
```

```python
y.head()
```

## Feature Scaling

```python
from sklearn.preprocessing import MinMaxScaler
mms=MinMaxScaler()
```

```python
X_mms=mms.fit_transform(X)
```

## Splitting the data into training set and testing set

```python
from sklearn.model_selection import train_test_split
```

```python
 X_train, X_test, y_train, y_test = train_test_split(X_mms, y, test_size=0.18, random_state=42)
```

## Linear Regression

```python
from sklearn.linear_model import LinearRegression
lin_regg=LinearRegression()
```

```python
lin_regg.fit(X_train,y_train)
```

```python
y_pred=lin_regg.predict(X_test)
```

```python
from sklearn import metrics
print('MAE:', metrics.mean_absolute_error(y_test, y_pred))
print('MSE:', metrics.mean_squared_error(y_test, y_pred))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```

```python
print("r2_score :" ,metrics.r2_score(y_test, y_pred))
```

```python
from sklearn.model_selection import cross_val_score
cv_score=cross_val_score(lin_regg,X,y,cv=5).mean()
cv_score
```

## Hyperparameter tuning
