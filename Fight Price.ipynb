## Handling The Duration column

```python
df["Duration"]
```

We can see here that all the rows in this column is not uniform like 2h 50m  some contains only h like 19h or some contains on ly m  lik 5m
so we wil make it uniform and then split it into hrs and min 

So need to split to hrs and min

```python
duration = list(df["Duration"])
for i in range(len(df["Duration"])):
    if len(duration[i].split()) != 2: ## This is varifies if the row contain both the hrs, min if one of them is misssing then this condition we get true
        if "h" in duration[i]: #If h is there in m in missing we will add 0m 
            duration[i] = duration[i].strip() + " 0m"   
        else:
            duration[i] = "0h " + duration[i] # here we are adding 
```

```python
            
duration_hours = []
duration_mins = []
for i in range(len(df["Duration"])):
    duration_hours.append(int(duration[i].split(sep = "h")[0]))    
    duration_mins.append(int(duration[i].split(sep = "m")[0].split()[-1]))
```

```python
df["Duration_hours"] = duration_hours
df["Duration_mins"] = duration_mins
```

```python
df.drop('Duration',axis=1,inplace=True)
```

```python
df.head()
```

```python
df.info()
```
